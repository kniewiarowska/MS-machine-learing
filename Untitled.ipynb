```python
# Analiza i modyfikacja danych
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Ewaluacja
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.metrics import precision_score, classification_report
from sklearn.metrics import recall_score, f1_score, accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
# machine learning
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import OneHotEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier

from imblearn.over_sampling import RandomOverSampler
from imblearn.over_sampling import SMOTE, ADASYN
from sklearn.metrics import balanced_accuracy_score
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from imblearn.ensemble import BalancedBaggingClassifier
from sklearn.model_selection import GridSearchCV

#Libraries
from sklearn import tree
from matplotlib import pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
#For work encode categorical atrubuts
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
#For do a best a work flow
from sklearn.pipeline import Pipeline
#Missing values
from sklearn.impute import SimpleImputer
    
#macro_precision = 0
#macro_recall = 0 
#macro_f1 = 0

models = [DecisionTreeClassifier(class_weight='balanced'), RandomForestClassifier(class_weight='balanced'), KNeighborsClassifier(), GaussianNB(),MLPClassifier(random_state=1, max_iter=300) ]
models_names = ['Tree', 'Random forest', 'K Neighbours', 'Naive Bayes', 'MLP', 'Normal']

def read_raw_data():
    df = pd.read_csv('data/final_data_raw2.csv')
    return df.drop(columns=df.columns[0], axis=1)


def read_grouped_data():
    df = pd.read_csv('data/final_data.csv')
    return df.drop(columns=df.columns[0], axis=1)


def calculate_feeling_rate_distribution(df):
    print('DATA:')
    count = df.count()[0]
    data = []
    numbers = df['feeling_rate'].unique()
    for i in numbers:
        occur = df['feeling_rate'].value_counts()[i]
        percent = (occur / count) * 100
        value = round(percent, 2)
        data.append(value)
        print(str(i) + ' ' + str(round(percent, 2)) + '%')
        
    fig, ax = plt.subplots()
    ax.pie(data, labels=numbers, autopct='%1.1f%%')


def print_df_information(df):
    print(df.head())
    print(df.info())
    print(df.describe())


def change_txt_data(df):
    print(df.describe(include=['O']))
    categoricals = list(df.select_dtypes(include=['O']).columns)
    encoder = OneHotEncoder(sparse_output=False)
    encoded = encoder.fit_transform(df[categoricals])
    train_ohe = pd.DataFrame(encoded, columns=np.hstack(encoder.categories_))
    df = pd.concat((df, train_ohe), axis=1).drop(categoricals, axis=1)
    print(df.head())
    return df

def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center')


def print_balanced_accuracy_for_models(balanced_accuracy, balanced_method_name):
    plt.bar(models_names, balanced_accuracy)
    addlabels(models_names,balanced_accuracy)

    #giving title to the plot
    plt.title(balanced_method_name)
     
    # giving X and Y labels
    plt.xlabel("Models")
    plt.ylabel("Balanced acccuracy [%]")
    plt.show

def print_balanced_accuracy_for_values(balanced_accuracy, title, balanced_method_names, y):
    plt.bar(balanced_method_names, balanced_accuracy)
    addlabels(balanced_method_names, balanced_accuracy)

    #giving title to the plot
    plt.title(title)
     
    # giving X and Y labels
    plt.xlabel('Methods')
    plt.ylabel(y)
    plt.show
```

```python

train_df = read_grouped_data()
train_df = change_txt_data(train_df)
Y = train_df['feeling_rate'].values
X = train_df.drop(['feeling_rate'], axis=1).values

sns.heatmap(train_df.corr(), annot=True)
plt.tight_layout()

calculate_feeling_rate_distribution(train_df)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=71830, stratify=Y)
```

```python
#Look for best tree parameters

def calculate_metrics_for_imbalanced_methods(X_train, Y_train, results_array, model, metrics):
    #print(str(model) + ' ' + str(imbalanced_method))
    model.fit(X_train, Y_train)
    Y_pred = model.predict(X_test)

    #Balanced acuurancy
    balanced_accurancy = balanced_accuracy_score(Y_test, Y_pred)
    results_array.append(round(balanced_accurancy, 2))
    #print(balanced_accurancy)

    # #Clasification report
    get_data_from_classification_report(Y_test, Y_pred, metrics)
    #print(classification_report(Y_test, Y_pred))

    #Confiusion Matrix
    #cm = confusion_matrix(Y_test, Y_pred, labels=model.classes_)
    #disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
    #disp.plot()

def get_data_from_classification_report(y_true, y_pred, metrics):
    report = classification_report(y_true, y_pred, output_dict=True)
   
    weight_accuracy = report['accuracy']
    macro_precision =  report['macro avg']['precision'] 
    macro_recall = report['macro avg']['recall']    
    macro_f1 = report['macro avg']['f1-score']

    metric = [weight_accuracy, macro_precision, macro_recall, macro_f1]
    metrics.append(metric)
    #print(report)

def normal(array, model, metric):
    calculate_metrics_for_imbalanced_methods(X_train, Y_train, array, 'Normal', model, metric)

def random_over_sampler(array, model, metric):
    ros = RandomOverSampler(random_state=0)
    X_resampled, Y_resampled = ros.fit_resample(X_train, Y_train)
    calculate_metrics_for_imbalanced_methods(X_resampled, Y_resampled, array, 'Random Over Sampler', model, metric)
 
def smote(array, model, metric):
    X_resampled, Y_resampled = SMOTE().fit_resample(X_train, Y_train)
    calculate_metrics_for_imbalanced_methods(X_resampled, Y_resampled, array, 'Smote', model, metric)

def adasyn(array, model, metric):
    X_resampled, Y_resampled = ADASYN().fit_resample(X_train, Y_train)
    calculate_metrics_for_imbalanced_methods(X_resampled, Y_resampled, array,'Adasyn', model, metric)

def balanced_bagging_classifier(array, model, metric):
    bbc = BalancedBaggingClassifier(base_estimator=model,
                                sampling_strategy='auto',
                                replacement=False,
                                random_state=0)
    calculate_metrics_for_imbalanced_methods(X_train, Y_train, array, 'BalancedBaggingClassifier', bbc, metric)
```

```python
def compare_imbalanced_methods_for_one_model(model, array, metrics):
    random_over_sampler(array, model, metrics)
    balanced_bagging_classifier(array, model, metrics)
    smote(array, model, metrics)
    adasyn(array, model, metrics)
    normal(array, model, metrics)
    

def print_comparison_plot(array, model_name, y):
    balanced_method_name = ['ROS', 'BBC', 'Smote',  'Adasyn', 'Normal']
    print_balanced_accuracy_for_values(array, model_name, balanced_method_name, y)    
```

```python
def caluclate_model_metrics_for_many_runs(model, accuracy, balanced, f1):    
    ros = 0
    bbc = 0
    smo = 0
    ada = 0
    normal = 0
    
    weight_accuracy_ros = 0
    weight_accuracy_bbc = 0
    weight_accuracy_smo = 0
    weight_accuracy_ada = 0
    weight_accuracy = 0

    f1_ros = 0
    f1_bbc = 0
    f1_smo = 0
    f1_ada = 0
    f1_normal = 0

    iter = 10
    for i in range(1, iter):
         array = []
         metrics = []
         compare_imbalanced_methods_for_one_model(model, array, metrics)
         ros = ros + array[0]
         bbc = bbc + array[1]
         smo = smo + array[2]
         ada = ada + array[3]
         normal = normal + array[4]
    
         weight_accuracy_ros = weight_accuracy_ros + metrics[0][0]
         weight_accuracy_bbc = weight_accuracy_bbc + metrics[1][0]
         weight_accuracy_smo = weight_accuracy_smo + metrics[2][0]
         weight_accuracy_ada = weight_accuracy_ada + metrics[3][0]
         weight_accuracy = weight_accuracy + metrics[4][0]

         f1_ros = f1_ros + metrics[0][3]
         f1_bbc = f1_bbc + metrics[1][3]
         f1_smo = f1_smo + metrics[2][3]
         f1_ada = f1_ada + metrics[3][3]
         f1_normal = f1_normal + metrics[4][3]

    accuracy.append(round(weight_accuracy_ros/iter, 2))
    accuracy.append(round(weight_accuracy_bbc/iter, 2))
    accuracy.append(round(weight_accuracy_smo/iter, 2))
    accuracy.append(round(weight_accuracy_ada/iter, 2))
    accuracy.append(round(weight_accuracy/iter, 2))
    
    balanced.append(round((ros/iter), 2))
    balanced.append(round((bbc/iter), 2))
    balanced.append(round((smo/iter), 2))
    balanced.append(round((ada/iter), 2))
    balanced.append(round((normal/iter), 2))

    f1.append(round((f1_ros/iter), 2))
    f1.append(round((f1_bbc/iter), 2))
    f1.append(round((f1_smo/iter), 2))
    f1.append(round((f1_ada/iter), 2))
    f1.append(round((f1_normal/iter), 2))
    
    print(accuracy)
    print(balanced)
    
         # print(weight_accuracy_bbc)
         # print(weight_accuracy_smo)
         # print(weight_accuracy_ada)
         
         # print(weight_accuracy_ro
         # macro_precision = macro_precision + metrics[0][1]
         # macro_recall = macro_recall + metrics[0][2]
         # macro_f1 = macro_f1 + metrics[0][3]
  
```
